eliminacion de ';' de archivo csv debido a que pyspark lo delimita con los archivo separados con coma
**delimiter=";"**
para quitar columnas dadas en _c0 _c1 ......
**header=True** 

```python
##read the dataset
df_pyspark=spark.read.format("csv").options(delimiter=";", header=True ).load("/content/drive/MyDrive/Pypask_Course/test3columns.csv")
```

#####cambio de dato de columna , dbido a que detecta todos los datos como string

```python
from pyspark.sql.functions import col
df_pyspark = df_pyspark.withColumn("exprience", col("exprience").cast("integer"))
df_pyspark = df_pyspark.withColumn("age", col("age").cast("integer"))
```

```python
df_pyspark.dtypes
```

```python
df_pyspark.describe().show()
```

#####Añadir columnas

```python
df_pyspark=df_pyspark.withColumn('Experience after 2 years',df_pyspark['exprience']+2)
```

```python
df_pyspark.show()
```

####eliminar clumna

```python
df_pyspark=df_pyspark.drop('Experience after 2 years')
```
