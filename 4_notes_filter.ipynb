```python
import pyspark
```

```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('dataframe').getOrCreate()
```

```python
##read the dataset
df_pyspark=spark.read.format("csv").options(delimiter=";", header=True ).load("/content/drive/MyDrive/Pypask_Course/test4Columns.csv")
```

```python
df_pyspark=df_pyspark.na.drop()
df_pyspark.show()
```

###FILTROS

```python
### Salario de la gente de menos 203
df_pyspark.filter("salary<203").show()
```

```python
df_pyspark.filter("salary<202").select(['name','age']).show()
```

```python
#varaias condiciones AND 
# AND &
# OR |
df_pyspark.filter((df_pyspark['salary']<3000) & 
                  (df_pyspark['salary']>=203)).show()
```

```python
#inversa
df_pyspark.filter(~(df_pyspark['Salary']<=203)).show()
```
